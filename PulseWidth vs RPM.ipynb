```python
%matplotlib inline
import matplotlib.pyplot as plt
import csv

plt.rcParams['figure.figsize'] = (15, 5)
plt.xkcd()
```

```python
headers = ['pulse_width', 'lft_rpm', 'rgt_rpm']

lft_registros = {}
rgt_registros = {}
with open('pulsewidth_vs_rpm.csv', 'r') as f:
    reader = csv.DictReader(f, fieldnames = headers)
    for n,reg in enumerate(reader):
        if n==0: continue
        lft_registros[float(reg['pulse_width'])] = float(reg['lft_rpm']) * -1
        rgt_registros[float(reg['pulse_width'])] = float(reg['rgt_rpm']) * -1
```

```python
plt.figure()
plt.xlim([1300,1700])
plt.ylim([-60, 60])
plt.grid(True)
plt.title('Left')
plt.plot(*zip(*[(k,v) for k,v in lft_registros.items()]), 'bD-')

plt.figure()
plt.xlim([1300,1700])
plt.ylim([-60, 60])
plt.grid(True)
plt.title('Right')
plt.plot(*zip(*[(k,v) for k,v in rgt_registros.items()]), 'bD-')
```
